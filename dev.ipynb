```python
import pandas as pd

df = pd.read_csv("Dicc.csv", sep=";")
```

```python
df.filter(regex="1\.\d").columns
```

```python
df.iloc[:, -23:] = df.filter(regex="\d\.\d").fillna(0).astype(int)
```

```python
# Create this dict as an ordered dict
int_competencias = {
    1: "A1",
    2: "A2",
    3: "B1",
    4: "B2",
    5: "C1",
    6: "C2",
}

def competencias_form_int_to_str(x):
    print(int_competencias.get(x))
        
competencias_form_int_to_str(2)
```

```python
df.filter(regex="\d\.\d").fillna(0).astype(int).applymap(int_competencias.get).fillna("")
```

```python
df["SINONIMOS"].fillna("").str.split(",").explode().str.strip()
```

```python
df["SINONIMOS"].fillna("").apply(lambda x: x.split(","))
```

```python
df["bag_of_words"] = df[["PALABRA_CLAVE","SINONIMOS"]].fillna("").apply(lambda x: x[0].split(",") + x[1].split(",")[], axis=1)
```

```python
# Select row if a word is in a list of words in a column
df["test"] = df["bag_of_words"].apply(lambda x: [x.strip() for x in x])
```

```python
"SIGAD-" in df["test"][2]
```

```python
# return a set of columns if a word is in a list of words in a column
df.loc[df["test"].apply(lambda x: "SIGAD-" in x), df.filter(regex="\d\.\d").columns]
```
