```python
import pandas as pd
from collections import OrderedDict

df = pd.read_csv("Dicc.csv", sep=";")
df.iloc[:, -23:] = df.filter(regex="\d\.\d").fillna(0).astype(int)
df["bag_of_words"] = (
        df[["PALABRA_CLAVE", "SINONIMOS"]]
        .fillna("")
        .apply(lambda x: x[0].lower().split(",") + x[1].lower().split(","), axis=1)
    )
df["bag_of_words"] = df["bag_of_words"].apply(
    lambda x: [x.strip() for x in x]
)

cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df.reindex(cols, axis = 1)
```

```python
df
```

```python
def filter_df(*args, df):
    dfs = []
    for word in args:
        word = word.strip().lower()
        df_filtered = df.loc[
            df["bag_of_words"].apply(lambda x: word in x),
            df.filter(regex="\d\.\d").columns,
        ]
        dfs.append(df_filtered)
        
    df= pd.concat(dfs)
    return df

lista_de_palabras_clave = [
    "GESTION ACADEMICA",
    "edición de audio",
]

df_filtrado = filter_df(*lista_de_palabras_clave, df=df)
```

```python
# compare values of diefferent cells of the same column and get the highest value

serie = df_filtrado[df_filtrado.filter(regex="\d\.\d").columns].max()
pd.DataFrame(serie).transpose()
```
